```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('flower.csv', names=['x1','x2','x3'])
```

```python

z_score = MinMaxScaler()
df = pd.DataFrame(z_score.fit_transform(df), columns=df.columns)
df
```

```python
c=np.array([df.iloc[0].values, df.iloc[7].values, df.iloc[13].values])
print(c)

for i in range(len(c)):
    df[f'c{i}_dist'] = df.iloc[:,:3].apply(lambda x: np.linalg.norm(x-c[i]), axis =1)
df['cluster'] = df.iloc[:,3:6].apply(lambda x: np.argmin(x),axis=1)
display(df)
```

```python
c = df.iloc[:,[0,1,2,6]].groupby('cluster').mean().values
print(c)
for i in range(len(c)):
    df[f'c{i}_dist'] = df.iloc[:,:3].apply(lambda x: np.linalg.norm(x-c[i]), axis =1)
df['cluster'] = df.iloc[:,3:6].apply(lambda x: np.argmin(x),axis=1)
display(df)
```
