```python
import cv2
import numpy as np
from PIL import Image
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
img_MtEverest= cv2.imread('/content/Mount_Everest.jpg')
plt.imshow(img_MtEverest[:,:,::-1])
plt.axis("off")
plt.show()
```

```python
#Add your code here!
operatedImage = cv2.cvtColor(img_MtEverest, cv2.COLOR_BGR2GRAY)
operatedImage = np.float32(operatedImage)

blocksize = [2,6,8]
ksize = [5,19,27]
plt.figure(figsize=(15,10))
for i in range(len(blocksize)):
  for j in range(len(ksize)):
    image = operatedImage.copy()
    img_Mt = img_MtEverest.copy()
    dest = cv2.cornerHarris(image, blocksize[i], ksize[j], 0.07)
    dest_dilate = cv2.dilate(dest, None)
    img_Mt[dest_dilate > 0.01 * dest_dilate.max()]=[0, 0, 255] #red color
    plt.subplot(len(blocksize),len(ksize),i*len(ksize)+j+1)
    plt.title('blocksize = '+str(blocksize[i])+', ksize = '+str(ksize[j]))
    plt.imshow(img_Mt[:,:,::-1])
    plt.axis("off")


```

```python
img1 = cv.imread('/content/Notebook01.jpeg')
plt.imshow(img1[:,:,::-1])
plt.show()
```

```python
img2 = cv.imread('/content/Notebook02.jpeg')
plt.imshow(img2[:,:,::-1])
plt.show()
```

```python
# Answer your code here

img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)

sift = cv2.SIFT_create()

kp, des = sift.detectAndCompute(img1,None)
img_with_keypoints_detail1 = cv2.drawKeypoints(
    img1,
    kp,
    None,
    flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS
)

kp, des = sift.detectAndCompute(img2,None)
img_with_keypoints_detail2 = cv2.drawKeypoints(
    img2,
    kp,
    None,
    flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS
)

plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.imshow(img_with_keypoints_detail1[:,:,::-1])
plt.subplot(1,2,2)
plt.imshow(img_with_keypoints_detail2[:,:,::-1])
plt.show()

```
