# Convert between color space, RGB, YCbCr, HSV, and show each channel in gray using matplotlib

```python
# Step 1: Load the image using OpenCV (default BGR format)
dog = cv2.imread('./Dogg.jpg')  # BGR format

# Step 2: Convert to RGB format
image_rgb  = cv2.cvtColor(dog, cv2.COLOR_BGR2RGB)

# Step 3: Display the RGB Image
plt.figure(figsize=(10, 8))
plt.imshow(image_rgb)  # Show RGB image
plt.title("Image in RGB")
plt.axis("off")  # Turn off axis
plt.show()
```

```python
# Split the channels for YCbCr
R, G, B = cv2.split(image_rgb)

# Display Y (Luminance) channel
plt.subplot(1, 3, 1)
plt.imshow(R, cmap='gray')
plt.title("Red")
plt.axis("off")

# Display Cb (Blue Chrominance) channel
plt.subplot(1, 3, 2)
plt.imshow(G, cmap='gray')
plt.title("Green")
plt.axis("off")

# Display Cr (Red Chrominance) channel
plt.subplot(1, 3, 3)
plt.imshow(B, cmap='gray')
plt.title("Blue")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
# Split the channels for YCbCr
R, G, B = image_rgb.copy(), image_rgb.copy(), image_rgb.copy()
R[:,:,1]=0
R[:,:,2]=0
G[:,:,0]=0
G[:,:,2]=0
B[:,:,0]=0
B[:,:,1]=0
# Display Y (Luminance) channel
plt.subplot(1, 3, 1)
plt.imshow(R)
plt.title("Red")
plt.axis("off")

# Display Cb (Blue Chrominance) channel
plt.subplot(1, 3, 2)
plt.imshow(G)
plt.title("Green")
plt.axis("off")

# Display Cr (Red Chrominance) channel
plt.subplot(1, 3, 3)
plt.imshow(B,)
plt.title("Blue")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
# Convert to YCbCr color space
image_ycbcr = cv2.cvtColor(dog, cv2.COLOR_BGR2YCrCb)

# Split the channels for YCbCr
Y, Cb, Cr = cv2.split(image_ycbcr)

# Display Y (Luminance) channel
plt.subplot(1, 3, 1)
plt.imshow(Y, cmap='gray')
plt.title("Y (Luminance)")
plt.axis("off")

# Display Cb (Blue Chrominance) channel
plt.subplot(1, 3, 2)
plt.imshow(Cb, cmap='gray')
plt.title("Cb (Blue Chrominance)")
plt.axis("off")

# Display Cr (Red Chrominance) channel
plt.subplot(1, 3, 3)
plt.imshow(Cr, cmap='gray')
plt.title("Cr (Red Chrominance)")
plt.axis("off")

plt.tight_layout()
plt.show()

```

```python
# Convert to YCbCr color space
image_ycbcr = cv2.cvtColor(dog, cv2.COLOR_BGR2YCrCb)

# Split the channels for YCbCr
Y, Cb, Cr = image_ycbcr.copy(), image_ycbcr.copy(), image_ycbcr.copy()
Y[:,:,1]=0
Y[:,:,2]=0
Cb[:,:,0]=0
Cb[:,:,2]=0
Cr[:,:,0]=0
Cr[:,:,1]=0
Y_rgb=cv2.cvtColor(Y, cv2.COLOR_YCrCb2RGB)
Cb_rgb=cv2.cvtColor(Cb, cv2.COLOR_YCrCb2RGB)
Cr_rgb=cv2.cvtColor(Cr, cv2.COLOR_YCrCb2RGB)

# Display Y (Luminance) channel
plt.subplot(1, 3, 1)
plt.imshow(Y_rgb)
plt.title("Y (Luminance)")
plt.axis("off")

# Display Cb (Blue Chrominance) channel
plt.subplot(1, 3, 2)
plt.imshow(Cb_rgb)
plt.title("Cb (Blue Chrominance)")
plt.axis("off")

# Display Cr (Red Chrominance) channel
plt.subplot(1, 3, 3)
plt.imshow(Cr_rgb)
plt.title("Cr (Red Chrominance)")
plt.axis("off")

plt.tight_layout()
plt.show()
```

#  Apply some linear function, adding or multiplying with constance, to each channel to observe the result

```python
# linear function: f(x)=ax+b
R, G, B = cv2.split(image_rgb)

# Apply linear transformations
a = 1  # Multiplication constant (scaling factor) ## contrast
b = 5   # Addition constant (offset) ## Brightness

R_transformed = np.clip(a * R + b, 0, 255).astype(np.uint8)
G_transformed = np.clip(a * G + b, 0, 255).astype(np.uint8)
B_transformed = np.clip(a * B + b, 0, 255).astype(np.uint8)

# Merge the transformed channels back into an image
image_transformed = cv2.merge([R_transformed, G_transformed, B_transformed])

# Visualize original and transformed images
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title("Original Image (RGB)")
plt.axis("off")

# Transformed Image
plt.subplot(1, 2, 2)
plt.imshow(image_transformed)
plt.title(f"Transformed Image (a={a}, b={b})")
plt.axis("off")

plt.tight_layout()
plt.show()
```

# Crop image using numpy index

```python
# Check shape
height, width, channels = image_rgb.shape
print(f"Height: {height}, Width: {width}, Channels: {channels}")
```

```python
# center crop
crop_height, crop_width = 500, 300  # cropped area height and width

start_row = (height - crop_height) // 2
end_row = start_row + crop_height
start_col = (width - crop_width) // 2
end_col = start_col + crop_width

print(f"Area to corp: Rows ({start_row}:{end_row}), Columns ({start_col}:{end_col})")

# crop
cropped_image = image_rgb[start_row:end_row, start_col:end_col]

plt.imshow(cropped_image)
plt.title("Cropped Image")
plt.axis("off")
plt.show()
```

```python
# Fixed Region Crop
# Step 1: Define cropping coordinates
start_row, end_row = 10, 300  # Define height range
start_col, end_col = 100, 600  # Define width range

# Step 2: Crop the image using NumPy slicing
cropped_image1 = image_rgb[start_row:end_row, start_col:end_col]

# Step 3: Display the cropped image
plt.imshow(cropped_image1)
plt.title("Cropped Image1")
plt.axis("off")
plt.show()

save_path = './cropped_image1.npy'  # 定义保存路径（子目录）
np.save(save_path, cropped_image1)  # 保存文件
print(f"File saved at: {save_path}")

# Convert cropped image back to BGR format for saving with OpenCV
cropped_image_bgr = cv2.cvtColor(cropped_image1, cv2.COLOR_RGB2BGR)

# Save the cropped image in .jpg format
jpg_save_path = './cropped_image1.jpg'
cv2.imwrite(jpg_save_path, cropped_image_bgr)
print(f"File saved as JPG at: {jpg_save_path}")


```
