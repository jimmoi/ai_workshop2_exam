3. Pandas ไม่สามารถอ่านข้อมูลได้

```
!gdown 17taZfXThpa3PXQeIR4EANfkGnf2US06s
```

```
data = pd.read_csv('universities-error.csv')
data.head(5)
```

อ่านไฟล์แบบปกติ เพื่อตรวจสอบว่าเกิดข้อผิดพลาดที่บรรทัดใด

```
err_file = open('universities-error.csv', 'r')
lines = err_file.readlines()
count = 0
for line in lines:
    count += 1
    if len(line.split(","))!=4:
      print("ERROR Line {}: {}".format(count, line.strip()))
      break
    else:
      print("Line {}: {}".format(count, line.strip()))
```

ใช้ parameter **on_bad_lines='skip'** เพื่อข้าม record ที่ไม่สามารถ parse ได้

สามารถระบุค่าต่อไปนี้ได้

*   'error', จะแสดง Exception เมื่อเกิด error
*   'warn', จะแสดงข้อความแจ้งเตือน และข้ามข้อมูลบรรทัดที่เกิด error
*   'skip', จะข้ามข้อมูลบรรทัดที่เกิด error โดยไม่มีการแจ้งเตือนใด ๆ


```
data = pd.read_csv('universities-error.csv', on_bad_lines='skip')
print(data.head(5))
```

# Exporting Pickle File

import library ที่เกี่ยวข้อง

```
import pickle
```

เปิดไฟล์สำหรับเขียนด้วยฟังก์ชัน open และใช้ mode เป็นการเขียนไฟล์ไบนารี (wb)

โหลดข้อมูลจากตัวแปรใส่ไฟล์ด้วย method dump() จาก library pickle

```
population_dict = {'California': 38332521, 'Texas': 26448193,
                   'New York': 19651127, 'Florida': 19552860,
                   'Illinois': 12882135}
area_dict = {'California': 423967, 'Texas': 695662, 'New York': 141297,
             'Florida': 170312, 'Illinois': 149995}
population = pd.Series(population_dict)
area = pd.Series(area_dict)
data = pd.DataFrame({'population': population, 'area': area})
density = pd.Series(data['population']/data['area'])

pkl_file = open('population.pkl', 'wb')
pickle.dump([data, density], pkl_file)
pkl_file.close()
```

# Importing Pickle File

เปิดไฟล์สำหรับอ่านด้วยฟังก์ชัน open และใช้ mode เป็นการอ่านไฟล์ไบนารี (rb)

อ่านข้อมูลใส่ตัวแปรด้วย method load() จาก library pickle

```
pkl_file = open('population.pkl', 'rb')
data = pickle.load(pkl_file)
for idx in range(len(data)):
  print("index {} => \n{}".format(idx, data[idx]))

pkl_file.close()
```

```
import h5py
```

เปิดไฟล์ HDF5 ด้วย h5py library

สำรวจ key (ทั้ง groups และ datasets) ด้วย method keys()

```
data = h5py.File('temperature.hdf5', 'r')
print(data.keys())
print([i for i in data.items()])
```

เข้าถึงข้อมูลผ่าน key ด้วย method get()

สำรวจสมาชิกในกลุ่มข้อมูลด้วย method items()

```
place_A = data.get('place_A')
print(place_A.keys())
print(place_A.items())
```

ใช้ list comprehension ในการตรวจสอบข้อมูลใน group place_A

```
print([i for i in place_A.items()])
```

ใช้ list comprehension ในการตรวจสอบ Attribute ของไฟล์ HDF5, groups, datasets ด้วย Attribute attrs

```
print([i for i in data.attrs])
print([i for i in place_A.attrs])
```

```
print("This file contains data from {} places.".format(data.attrs['num_of_place']))
print("Title of this dataset: {}".format(data.attrs['title']))
print("Location of Place A at LAT {}, LONG {}".format(place_A.attrs['lat'],place_A.attrs['long']))
```

ใช้ method get() ในการเข้าถึงแบบลำดับชั้นได้

```
place_A_summary = data.get('place_A/summary')
print([i for i in place_A_summary.items()])
```

นำค่ามาใช้ด้วยการเข้าถึงผ่าน index

```
avg_weeks = place_A_summary.get('weeks')
avg_weeks[()]
```

```
avg_weeks[0]
```

หรือการแปลงค่าเป็น Numpy array

```
np.array(avg_weeks)
```
