# Image Thresholding



##Simple Thresholding (global value as a threshold)



*   If the pixel value is smaller than the threshold, it is set to 0, otherwise it is set to a maximum value.
*   All simple thresholding types are:
 * cv.THRESH_BINARY
 * cv.THRESH_BINARY_INV
 * cv.THRESH_TRUNC
 * cv.THRESH_TOZERO
 * cv.THRESH_TOZERO_INV
 Overview more in this link [here!](https://docs.opencv.org/3.4/d7/d1b/group__imgproc__misc.html#ggaa9e58d2860d4afa658ef70a9b1115576a147222a96556ebc1d948b372bcd7ac59)

* **cv.threshold** is used to apply the thresholding.
 * Example : cv.threshold(img,127,255,cv.THRESH_BINARY_INV)

   * 'img' (first argument) = source of image (grayscale image)
   * '127' (second argument) = is the threshold value which is used to classify the pixel values.
   * '255' (third argument) = is the maximum value which is assigned to pixel values exceeding the threshold.
   * 'cv.THRESH_BINARY_INV' (forth argument) = is placed for defining different types of thresholding.

### cv.THRESH_BINARY

```python
ret,thresh1 = cv.threshold(img,127,255,cv.THRESH_BINARY)
titles = ['Original Image','BINARY']
images = [img, thresh1]
for i in range(len(images)):
  plt.subplot(1,2,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
  plt.title(titles[i])
  plt.xticks([]),plt.yticks([])
plt.show()
```



 ## Adaptive Thresholding** (different thresholds for different regions)


*   In case an image has different lighting conditions in different areas.
*   The adaptiveMethod decides how the threshold value is calculated:

 * cv.ADAPTIVE_THRESH_MEAN_C: The threshold value is the mean of the neighbourhood area minus the constant C.
 * cv.ADAPTIVE_THRESH_GAUSSIAN_C: The threshold value is a gaussian-weighted sum of the neighbourhood values minus the constant C.





```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
!gdown '1bjJ38_JCOuYGmDcC9Addkl_ZTWkh8MX6'
```

```python
img = cv.imread('sudoku.png', cv.IMREAD_GRAYSCALE)
assert img is not None, "file could not be read, check with os.path.exists()"
img_blur = cv.medianBlur(img,5)
```

```python
cv2_imshow(img)
```

```python
cv2_imshow(img_blur)
```

```python
titles = ['Original Image', 'Global Thresholding (v = 127)']
ret,th1 = cv.threshold(img,127,255,cv.THRESH_BINARY)

images = [img, th1]
for i in range(len(images)):
    plt.subplot(1,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

 ### ADAPTIVE_THRESH_MEAN_C function
> the threshold value t(x,y)
 is a mean of the (blockSize X blockSize)
 neighborhood of
 minus C



```python
titles = ['Original Image', 'Adaptive Mean Thresholding']

th2 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_MEAN_C,\
            cv.THRESH_BINARY,11,2)

images = [img, th2]
for i in range(len(images)):
    plt.subplot(1,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

 ### ADAPTIVE_THRESH_GAUSSIAN_C function
> the threshold value T(x,y)
 is a weighted sum (cross-correlation with a Gaussian window) of the (blockSize X blockSize)
 neighborhood of (x,y)
 minus C . The default sigma (standard deviation) is used for the specified blockSize .



```python
titles = ['Original Image', 'Adaptive Gaussian Thresholding']

th3 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv.THRESH_BINARY,11,2)
images = [img, th3]
for i in range(len(images)):
    plt.subplot(1,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

 ### Showing all result of adaptive thresholding

```python
titles = ['Original Image', 'Global Thresholding (v = 127)',
            'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
ret,th1 = cv.threshold(img,127,255,cv.THRESH_BINARY)
th2 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_MEAN_C,\
            cv.THRESH_BINARY,11,2)
th3 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv.THRESH_BINARY,11,2)
images = [img, th1, th2, th3]
for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```



## Otsu's Binarization (global thresholding)
*   This method not only avoids having to choose a value and determines it automatically but also determines an optimal global threshold value from the image histogram (consist of two peaks). **A good threshold would be in the middle of those two values.




```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
!gdown '1Ac7Bj-ohEFFVdcpUCRk4GRqkqF3n_RMt'
```

```python
img = cv.imread('noisy2.png', cv.IMREAD_GRAYSCALE)
assert img is not None, "file could not be read, check with os.path.exists()"
```

### global thresholding with a value of 127 is applied.

```python
# global thresholding
ret1,th1 = cv.threshold(img,127,255,cv.THRESH_BINARY)
```

### Otsu's thresholding is applied directly.

```python
# Otsu's thresholding
ret2,th2 = cv.threshold(img,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
```

### The image is first filtered with a 5x5 gaussian kernel to remove the noise, then Otsu thresholding is applied.

```python
# Otsu's thresholding after Gaussian filtering
blur = cv.GaussianBlur(img,(5,5),0)
ret3,th3 = cv.threshold(blur,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
```

###Plot all the images and their histograms

```python
# plot all the images and their histograms
images = [img, 0, th1,
          img, 0, th2,
          blur, 0, th3]
titles = ['Original Noisy Image','Histogram of Original','Global Thresholding (v=127)','Histogram of Thresholding',
          'Original Noisy Image','Histogram of Original',"Otsu's Thresholding",'Histogram of Thresholding',
          'Gaussian filtered Image','Histogram of Original',"Otsu's Thresholding",'Histogram of Thresholding',]

plt.figure(figsize=(12, 9))

for i in range(3):
    plt.subplot(3,4,i*4+1),plt.imshow(images[i*3],'gray')
    plt.title(titles[i*4]), plt.xticks([]), plt.yticks([])
    plt.subplot(3,4,i*4+2),plt.hist(images[i*3].ravel(),256)
    plt.title(titles[i*4+1]), plt.xticks([]), plt.yticks([])
    plt.subplot(3,4,i*4+3),plt.imshow(images[i*3+2],'gray')
    plt.title(titles[i*4+2]), plt.xticks([]), plt.yticks([])
    plt.subplot(3,4,i*4+4),plt.hist(images[i*3+2].ravel(),256)
    plt.title(titles[i*4+3]), plt.xticks([]), plt.yticks([])
plt.show()
```
